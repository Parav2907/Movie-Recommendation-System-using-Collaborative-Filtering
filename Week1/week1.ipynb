```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('week1_data.csv')
```

```python
action = df[df['genres'].str.contains('"Action"', case=False, na=False)].count().iloc[0]
```

```python
categories=['Action', 'Adventure', 'Fantasy', 'Science Fiction', 'Crime', 'Drama','Thriller', 'Animation', 'Family', 'Western', 'Comedy', 'Romance', 
 'Horror', 'Mystery', 'History', 'War', 'Music', 'Documentary','Foreign', 'TV Movie']
```

```python
value=[]
for gen in categories:
    val= df[df['genres'].str.contains(gen, case=False, na=False)].count().iloc[0]
    value.append(val)
```

```python
plt.bar(categories, value)
plt.title('Number of movies per Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')
plt.show()
```
